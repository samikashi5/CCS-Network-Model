This notebook uses data in data/sources.csv and code in model/ccs_network_optimization.py.

Before running: fill all NEED INFO values in the Parameters class inside model/ccs_network_optimization.py.

```python
from pathlib import Path
import sys

import matplotlib.pyplot as plt

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from model.ccs_network_optimization import Parameters, build_and_solve_network, load_sources, save_results
```

```python
data_path = ROOT / "data" / "sources.csv"
sources = load_sources(data_path)
sources
```

```python
params = Parameters(
    discount_rate=0.10,  # NEED INFO
    trucking_cost_usd_per_ton_km=0.08,  # NEED INFO
    pipeline_cost_usd_per_ton_km=0.03,  # NEED INFO
    storage_cost_usd_per_ton=15.0,  # NEED INFO
    sink_capacity_tpy=2_000_000.0,  # NEED INFO
    capture_target_tpy=None,  # NEED INFO
)
params
```

```python
results = build_and_solve_network(sources, params)
results["summary"]
```

```python
allocation = results["allocation"]
allocation
```

```python
plot_df = allocation.copy()
plot_df["flow_total_tons"] = plot_df["flow_pipeline_tons"] + plot_df["flow_truck_tons"]
ax = plot_df.set_index("source")[["flow_pipeline_tons", "flow_truck_tons"]].plot(kind="bar", figsize=(10, 5))
ax.set_title("Allocated CO2 Flow by Source and Mode")
ax.set_xlabel("Source")
ax.set_ylabel("Flow (tons/year)")
plt.tight_layout()
plt.show()
```

```python
save_results(results)
print("Saved outputs to:", ROOT / "outputs")
```
